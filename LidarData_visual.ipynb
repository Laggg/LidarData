# I Вариант

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def see_plot(scan):
    fig = plt.figure(1, figsize=(15,10))
    ax = fig.add_subplot(111, projection='3d', elev=50, azim=-20)
    ax.scatter(scan[:,:,0],scan[:,:,1],scan[:,:,2],c=scan[:,:,3],cmap='plasma',marker='o')
    plt.show()
#===========================================================================================

filepath = 'C:\\Users\\aaklo\\Downloads\\mipt_people3.bin'
scans = np.fromfile(filepath, dtype=np.float32) #
scans = scans.reshape((-1, 64, 2048, 4))

scan = np.clip(scans[0], -20, 20)
see_plot(scan)
```

```python
import numpy as np
import open3d as o3d # https://anaconda.org/open3d-admin/open3d

data = np.fromfile('C:\\Users\\aaklo\\Downloads\\mipt_people3.bin')
pc = o3d.geometry.PointCloud()
pc.points = o3d.utility.Vector3dVector(data.reshape((-1, 64, 2048, 4))[0].reshape(-1, 4)[:, :3])

o3d.visualization.draw_geometries([pc], width=1500, height=700)
```
